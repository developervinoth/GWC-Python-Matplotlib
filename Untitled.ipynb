```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
 y= x**2
```

```python
y
```

# Functional Method in Matplot Lib

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-.')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

# Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X_Label')
axes.set_ylabel('Y_Label')
```

```python
newFig = plt.figure()
axes1 = newFig.add_axes([0.1,0.1,0.8,0.8])
axes2 = newFig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
axes1.set_title('TitleForAxes1')
axes2.set_title('TitleForAxes2')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =2)
for i in axes:
    i.plot(x,y)
```
