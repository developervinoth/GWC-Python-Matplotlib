```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,10)
```

```python
y = x**2
```

# Functional Method

```python
plt.plot(x,y,'r-.')
plt.xlabel('This is a X_lable')
plt.ylabel('This is a Y_lable')
plt.title('This this a Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x)
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('This is a X label')
axes.set_ylabel('This is Y Label')
axes.set_title('This is a Title')
```

```python
figNew = plt.figure()
axes1 = figNew.add_axes([0.1,0.1,0.8,0.8])
axes2 = figNew.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x)
axes1.set_title('This the axes One')
axes2.set_title('This the axes Two')
axes1.set_xlabel('Xlable for Axes1')
axes1.set_ylabel('Ylable for Axes1')
```

```python
fig, axes = plt.subplots(nrows =1, ncols = 2)
for i in axes:
    i.plot(x,y)
```

```python
figWithDpi = plt.figure(dpi = 150)
plot1 = figWithDpi.add_axes([0.1,0.1,0.8,0.8])
plot1.plot(x,y)

figWithDpi.savefig('TestImage.png', dpi = 200)
```
